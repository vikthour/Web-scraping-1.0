Using while and for loop to replace repititive blocks of code.

```python
url = 'https://www.vanguardngr.com/category/more/metro/page/'
news_heading = []
time_strip = []
i = 1

while i <= 300:
    url = url[:53] + str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    news_link = soup.findAll('h2', class_="entry-title")
    for j in np.arange(10):
        time_link = soup.findAll('time', class_ = "entry-date published")
        link = news_link[j].text.strip()
        time = time_link[j].text.strip()
        news_heading.append(link)
        time_strip.append(time)
        
    i += 1
```

```python
df_news = pd.DataFrame({'Headlines': news_heading, 'Date': time_strip}, index = np.arange(3000))
df_news.head()
```

Generate a list of keywords needed for the search.

```python
keywords = ['crash', 'crush', 'ghast', 'tanker', 'truck', 'accident', 'collision', 'wreck', 'reck', 'drive', 'hit']
```

Use for loop to iterate over entire dataset while the keywords to determine the viable news.

```python
accident = []
date = []
for i in range(len(df_news)):
    for key in keywords:
        if key in df_news.Headlines[i]:
            news = df_news.Headlines[i]
            time = df_news.Date[i]
            accident.append(news)
            date.append(time)
```

Convert the filtered dataset into a DataFrame.

```python
Road_accident = pd.DataFrame({'Headline': accident, 'Date': date}, index =None)
Road_accident.head(3)
```
